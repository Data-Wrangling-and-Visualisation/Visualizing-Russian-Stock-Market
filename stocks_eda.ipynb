```python
import pandas as pd
import matplotlib.pyplot as plt
import mplfinance as mpf
```

```python
# dataframe with stocks data
df_stocks = pd.read_json('./Database/combined_data_stock.json', lines=True)
df_stocks['begin'] = pd.to_datetime(df_stocks['begin'])

df_stocks
```

<h4>Visualization of the data of the particular company (with ticker RNFT) as an example</h4>

```python
# choosing the necessary ticker
ticker1 = 'RNFT'
filtered_data1 = df_stocks[df_stocks['STOCK_TICK'] == ticker1]
fig, axs = plt.subplots(3, 2, figsize=(20, 10))

# plotting graph of open prices
axs[0, 0].plot(filtered_data1['begin'], filtered_data1['open'], color='blue')
axs[0, 0].set_title('Open Prices')
axs[0, 0].set_xlabel('Date')
axs[0, 0].set_ylabel('Price')
axs[0, 0].grid()

# plotting graph of close prices
axs[0, 1].plot(filtered_data1['begin'], filtered_data1['close'], color='green')
axs[0, 1].set_title('Close Prices')
axs[0, 1].set_xlabel('Date')
axs[0, 1].set_ylabel('Price')
axs[0, 1].grid()

# plotting graph of low prices
filtered_data1['low'] = filtered_data1['high']
axs[1, 0].plot(filtered_data1['begin'], filtered_data1['low'], color='orange')
axs[1, 0].set_title('Low Prices')
axs[1, 0].set_xlabel('Date')
axs[1, 0].set_ylabel('Price')
axs[1, 0].grid()

# plotting graph of high prices
axs[1, 1].plot(filtered_data1['begin'], filtered_data1['high'], color='red')
axs[1, 1].set_title('High Prices')
axs[1, 1].set_xlabel('Date')
axs[1, 1].set_ylabel('Price')
axs[1, 1].grid()

# plotting graph of values
axs[2, 0].plot(filtered_data1['begin'], filtered_data1['value'], color='purple')
axs[2, 0].set_title('Values')
axs[2, 0].set_xlabel('Date')
axs[2, 0].set_ylabel('Price')
axs[2, 0].grid()

# plotting graph of volumes
axs[2, 1].plot(filtered_data1['begin'], filtered_data1['volume'], color='magenta')
axs[2, 1].set_title('Volumes')
axs[2, 1].set_xlabel('Date')
axs[2, 1].set_ylabel('Number of shares')
axs[2, 1].grid()

fig.suptitle(f"Information about {ticker1} stocks")
plt.tight_layout()
plt.show()
```

<h4>Candlestick graphs for the RNFT data</h4>


Here, green means that the closing price was higher than the opening price; red means the opposite. 

```python
# setting the column "begin" as index in RNFT ticker dataframe
filtered_data1.set_index('begin', inplace=True)
```

```python
# selecting monthly data of RNFT
monthly_data = filtered_data1.resample('M').agg({
    'open': 'first',
    'high': 'max',
    'low': 'min',
    'close': 'last',
    'volume': 'sum'
})

# plotting the chart
mpf.plot(
    monthly_data,
    type='candle',
    title=f'Monthly Candlestick Chart for {ticker1}',
    style='charles',
    figsize=(14, 7) 
)
```

```python
# selecting yearly data of RNFT
yearly_data = filtered_data1.resample('Y').agg({
    'open': 'first',
    'high': 'max',
    'low': 'min',
    'close': 'last',
    'volume': 'sum'
})

# plotting the chart
mpf.plot(
    yearly_data,
    type='candle',
    title=f'Yearly Candlestick Chart for {ticker1}',
    style='charles',
    figsize=(14, 7) 
)
```

<h4>Comparison between the companies</h4>

```python
# dataframe with mean of values for different companies
mean_values = df_stocks.groupby('STOCK_TICK')['value'].mean().reset_index()

# plotting the graph
plt.figure(figsize=(10, 6))
plt.bar(mean_values['STOCK_TICK'], mean_values['value'], color='purple')
plt.title('Mean Value of Different Companies')
plt.xlabel('Company Ticker')
plt.ylabel('Mean Value (â‚½)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# dataframe with mean of volumes for different companies
mean_values = df_stocks.groupby('STOCK_TICK')['volume'].mean().reset_index()

# plotting the graph
plt.figure(figsize=(10, 6))
plt.bar(mean_values['STOCK_TICK'], mean_values['volume'], color='magenta')
plt.title('Mean Volume of Different Companies')
plt.ylabel('Mean Volume')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
